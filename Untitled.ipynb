```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Superstore_sales.csv',na_values=['Missing','n/a','N/A','NA',''])
```

```python
train = df.sample(frac=0.8,random_state=123)
train
```

```python
train.columns.unique()
```

```python
train['Ship Mode'].unique()
```

```python
def ship_mode(x):
    if x=='Standard Class':
        return 0
    elif x=='Second Class':
        return 1
    elif x=='First Class':
        return 2
    if x=='Same Day':
        return 4
```

```python
train['shipmode_bkt']= train['Ship Mode'].map(ship_mode)
```

```python
train['shipmode_bkt'].head()
```

```python
pd.crosstab(train['Sales'],train['shipmode_bkt'],margins=True)/train.shape[0]
```

```python
train['Segment'].unique()
```

```python
def seg_mode(x):
    if x=='Corporate':
        return 0
    elif x=='Consumer':
        return 1
    elif x=='Home Office':
        return 2
```

```python
train['segment_bkt']=train['Segment'].map(seg_mode)
```

```python
train['segment_bkt'].head()
```

```python
train.head()
```

```python
train['Region'].unique()
```

```python
def state(x):
    if x=='South':
        return 0
    elif x=='West':
        return 1
    elif x=='East':
        return 2
    if x=='Central':
        return 4
```

```python
train['region_bkt']= train['Region'].map(state)
```

```python
train['region_bkt'].head()
```

```python
train.isnull().sum()
```

```python
train['Country'].unique()
```

```python
train['Sub-Category'].unique()
```

```python
def category(x):
    if x=='Furniture':
        return 0
    elif x=='Office Supplies':
        return 1
    elif x=='Technology':
        return 2
```

```python
train['category_bkt']= train['Category'].map(category)
train['category_bkt'].head()
```
